```python

# Create a sky-blue background
img = np.full((500, 500, 3), (235, 206, 135), dtype=np.uint8) 

cv2.line(img, (0, 240), (500, 240), (255, 255, 255), 2) # Road Lines

cv2.line(img, (0, 380), (500, 380), (255, 255, 255), 2) # Road Lines


cv2.rectangle(img, (50, 175), (120, 240), (238, 130, 238), -1) # School
 
cv2.putText(img, "SCHOOL", (55, 192), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)

cv2.rectangle(img, (180, 175), (250, 240), (0, 165, 255), -1) # Bank

cv2.putText(img, "BANK", (192, 200), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)

cv2.rectangle(img, (0, 380), (500, 500), (0, 255, 0), -1) # Grass


cv2.rectangle(img, (50, 400), (120, 460), (42, 42, 165), -1)  # Wine shop


cv2.putText(img, "WINE SHOP", (55, 420), cv2.FONT_HERSHEY_SIMPLEX, 0.35, (255, 255, 255), 1, cv2.LINE_AA)


cv2.rectangle(img, (350, 170), (360, 240), (42, 42, 165), -1) # Tree 1

cv2.circle(img, (355, 170), 20, (0, 255, 0), -1)

cv2.rectangle(img, (300, 170), (310, 240), (42, 42, 165), -1) # Tree 2

cv2.circle(img, (305, 170), 20, (0, 255, 0), -1)

cv2.putText(img, 
        "CONSUMES ALCOHOL...", 
        (140, 490), 
        cv2.FONT_HERSHEY_SIMPLEX, 
        0.5,  # Font scale
        255,  # Color (white)
        1,    # Thickness
        cv2.LINE_AA
    )



cv2.rectangle(img, (0, 240), (500, 380), (0, 0, 0), -1) # Road

# Draw the dashed road lines (white dashes)
for i in range(0, 500, 50):
    cv2.line(img, (i, 310), (i + 20, 310), (255, 255, 255), 2)

# Draw the vehicle 
cv2.rectangle(img, (25, 250), (100, 295), (0, 0, 255), -1)  # Filled slanted rectangle
cv2.circle(img, (40, 295), 8, (255, 255, 255), -1)  # Left wheel
cv2.circle(img, (85, 295), 8, (255, 255, 255), -1)  # Right wheel
cv2.rectangle(img, (65, 255), (95, 275), (255, 255, 255), 2) # Window


'''top_left = (385, 220)
bottom_right = (460, 265)

# Calculate the rectangle's center
center_x = (top_left[0] + bottom_right[0]) // 2
center_y = (top_left[1] + bottom_right[1]) // 2
center = (center_x, center_y)

angle = 30  # Adjust the angle as needed

# Get the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Draw the original rectangle
rectangle = np.array([[top_left[0], top_left[1]],[bottom_right[0], top_left[1]],[bottom_right[0], bottom_right[1]],[top_left[0], bottom_right[1]]], dtype=np.float32)

# Rotate the rectangle by applying the rotation matrix to each corner
rotated_rectangle = cv2.transform(np.array([rectangle]), rotation_matrix)[0]

# Convert the rotated rectangle coordinates to integers
rotated_rectangle = np.int32(rotated_rectangle)

# Draw the rotated rectangle on the image
#cv2.polylines(img, [rotated_rectangle], isClosed=True, color=(0, 0, 255), thickness = -1)
cv2.fillPoly(img, [rotated_rectangle], (0, 0, 255))'''

#cv2.circle(img, (428, 265), 8, (255, 255, 255), -1)  # Left wheel
#cv2.circle(img, (454, 250), 8, (255, 255, 255), -1)  # Right wheel


# Draw the moon (yellow circle)
cv2.circle(img, (450, 50), 20, (0, 215, 255), -1)

# Draw utility poles (brown lines)
cv2.rectangle(img, (450, 170), (460, 240), (42, 42, 165), -1)
cv2.rectangle(img, (440, 175), (470, 180), (0, 0, 0), -1)

'''cv2.rectangle(img, (450, 195), (460, 240), (42, 42, 165), -1)
cv2.rectangle(img, (440, 201), (470, 206), (0, 0, 0), -1)'''

x1, y1 = -1, -1
drawing = False

def mouse_tracking(event, x, y, flags, param):
    global x1, y1, drawing
    # Left button pressed
    if event == cv2.EVENT_LBUTTONDOWN:
        x1, y1 = x, y
        drawing = True
    # Left button released
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.line(img, (x1, y1), (x, y), (255, 255, 255), 2)  # Draw white line
    # Right button pressed
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 8, (255, 255, 255), -1)  # Draw white circle

cv2.namedWindow("animation")
cv2.setMouseCallback("animation", mouse_tracking)
i = 2
while True:
    cv2.imshow("animation", img)
    key = cv2.waitKey(1) & 255
    if key == ord("s"):
        cv2.imwrite(fr"C:\Users\hp\OneDrive\Desktop\ML\Project 1 Animation\Images\{i}.jpg", img)
        print("Image captured")
    elif key == ord("q"): 
        break

cv2.destroyAllWindows()
```

### Saving Video to a Path

```python
#Video properties
output_path = r"C:\Users\hp\OneDrive\Desktop\ML\Project 1 Animation & GIF\Animation Images\animation_output.avi"
frame_width = 640  # Set based on your image dimensions
frame_height = 480  # Set based on your image dimensions
frame_rate = 1  # 1 frame per second (based on your cv2.waitKey(1000))

#Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Choose codec (e.g., 'XVID' for .avi format)
out = cv2.VideoWriter(output_path, fourcc, frame_rate, (frame_width, frame_height))

for image in range(1, 17):
    img = cv2.imread(fr"C:\Users\hp\OneDrive\Desktop\ML\Project 1 Animation & GIF\Animation Images\{image}.jpg")

    if img is None:
        print(f"Image not found, skipping.")
        continue

    # Resize the image if necessary (to match frame dimensions)
    img = cv2.resize(img, (frame_width, frame_height))

#Display the image
    cv2.imshow("animations", img)
    cv2.waitKey(1000)

#Write the frame to the video
    out.write(img)

#Release the VideoWriter and close all OpenCV windows
out.release()
cv2.destroyAllWindows()

print(f"Video saved to {output_path}")
```
